```python
# ============================================
# FASTEST METHOD: aria2c with 16 connections
# ============================================

# Install aria2c
!apt-get install -y aria2 > /dev/null 2>&1

import subprocess
import requests
import os

def fast_download_zenodo(record_id, output_dir='./data', connections=16):
    """Ultra-fast download using aria2c"""

    # Get record
    print(f"Fetching record {record_id}...")
    r = requests.get(f"https://zenodo.org/api/records/{record_id}")
    record = r.json()

    print(f"\nTitle: {record['metadata']['title']}")
    print(f"Files: {len(record['files'])}")
    print(f"Using aria2c with {connections} parallel connections\n")

    os.makedirs(output_dir, exist_ok=True)

    # Download each file
    for i, f in enumerate(record['files'], 1):
        filename = f['key']
        url = f['links']['self']
        size_gb = f['size'] / (1024**3)

        print(f"[{i}/{len(record['files'])}] {filename} ({size_gb:.2f} GB)")

        output_path = os.path.join(output_dir, filename)

        # Skip if exists
        if os.path.exists(output_path) and os.path.getsize(output_path) == f['size']:
            print("✓ Already exists\n")
            continue

        # Download with aria2c
        cmd = [
            'aria2c',
            f'--max-connection-per-server={connections}',
            f'--split={connections}',
            '--min-split-size=1M',
            '--file-allocation=none',
            '--continue=true',
            f'--dir={output_dir}',
            f'--out={filename}',
            url
        ]

        subprocess.run(cmd)
        print()

    print(f"✓ Downloaded to {output_dir}/")
    return output_dir

# USE IT:
fast_download_zenodo(17668236, './data', connections=16)
```

```python
from mmsdk.mmdatasdk import mmdataset
recipe = {
    "COVAREP": "/content/CMU-MultimodalSDK/data/COVAREP.csd",
    "FACET": "/content/CMU-MultimodalSDK/data/VisualFacet42.csd",
    "WORDVEC": "/content/CMU-MultimodalSDK/data/TimestampedWordVectors.csd",
    "Labels": "/content/CMU-MultimodalSDK/data/Labels.csd"
}
dataset = mmdataset(recipe)
```

```python
labels_key = dataset["Labels"].keys()
```

```python
labels = dataset["Labels"]
audios = dataset["COVAREP"]
visuals = dataset["FACET"]
wordvecs = dataset["WORDVEC"]
```

```python
def is_inside_any(intervals, query):
    q_start, q_end = query

    # Vectorized condition:
    inside = (q_start >= intervals[:, 0]) & (q_end <= intervals[:, 1])

    return inside.any()
def extract_interval_features(feat_intervals, feat_values, label_interval):
    """
    feat_intervals : (T, 2) array of [start, end]
    feat_values    : (T, D) feature matrix
    label_interval : (2,) array [label_start, label_end]
    """
    ls, le = label_interval  # label start, label end

    # Vectorized mask: frame interval fully inside the label interval
    mask = (feat_intervals[:, 0] >= ls) & (feat_intervals[:, 1] <= le)

    # Extract only those frames
    return {"intervals": np.array(feat_intervals[mask]), "features": np.array(feat_values[mask])}
```

```python
from mmsdk import mmdatasdk
import numpy as np
data = {}

for k in labels_key:
  if k not in audios.keys() or k not in visuals.keys() or k not in wordvecs.keys():
    labels.data.pop(k, None)
c = 1
for label_key in labels_key:
  data_point = []
  for i in range(len(labels.data[label_key]["intervals"])):
    label_interval = labels.data[label_key]["intervals"][i]
    label_features = labels.data[label_key]["features"][i]
    # handle audio
    audio_feat = extract_interval_features(audios.data[label_key]["intervals"], audios.data[label_key]["features"], label_interval)
    # handle visual
    visual_feat = extract_interval_features(visuals.data[label_key]["intervals"], visuals.data[label_key]["features"], label_interval)
    # handle text
    wordvec_feat = extract_interval_features(wordvecs.data[label_key]["intervals"], wordvecs.data[label_key]["features"], label_interval)
    data_point.append({"Labels": {"intervals": label_interval, "features": label_features}, "COVAREP": audio_feat, "FACET": visual_feat, "WORDVEC": wordvec_feat})
  data[label_key] = data_point
  print(label_key, c)
  c += 1
```
