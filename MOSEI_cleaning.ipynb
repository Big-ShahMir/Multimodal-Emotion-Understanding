Since the Dataset is too large, we would want to truncate them first before aligning those visuals and text. We will do it using labels then use the key from label to extract for what we need. While extracting, we try to balanced the number of emotions.

````python
import math
````

````python
# 'dimension names': ['sentiment','happy','sad','anger','surprise','disgust','fear'],
emotions = ['happy','sad','anger','surprise','disgust','fear']
def is_equal(a, b):
  return b >= 0.1 and abs(a - b) < 1e-5
def extract_emotion_sentiment(arr):
  np_arr = np.array(arr)[1:]
  indices = np.where(is_equal(np_arr, np_arr.max()))
  indication_vector = np.zeros(6, dtype=int)
  indication_vector[indices] = 1
  return indication_vector
````

````python
# function to get the emotion and sentiment
def get_emotion(label_key):
  label = labels.data[label_key]
  features = np.array(label["features"])
  intervals = np.array(label["intervals"])
  emotions = np.zeros(6, dtype=int)
  for i in range(min(len(features) , len(intervals))):
    emotions += extract_emotion_sentiment(features[i])
  return emotions
````

````python
key_list = list(labels.data.keys())
emotion_sum = []
for x in key_list:
  emotion_sum.append(get_emotion(x))
print(np.array(emotion_sum))
````

````python
col_sum = np.sum(emotion_sum, axis=0)
print(col_sum)
````

````python
balance_emotion = []
balance_key = []
for x in range(len(emotion_sum)):
  if (emotion_sum[x][-3] > 0 or emotion_sum[x][-1] > 0):
    balance_emotion.append(emotion_sum[x])
    balance_key.append(key_list[x])
````

````python
print(np.sum(balance_emotion, axis=0))
print(balance_key)
print(len(balance_key))
````

````python
# ============================================
# COMPLETE UPLOAD SCRIPT - PASTE THIS IN COLAB
# ============================================

!pip install requests -q

import requests
import os
import json
from getpass import getpass

# Get your Zenodo token
# Corrected: Use a proper prompt for getpass, and ensure it gets a valid token.
# Alternatively, for Colab, you can use google.colab.userdata.get('ZENODO_TOKEN')
# or simply assign your token directly for testing purposes.
ZENODO_TOKEN = getpass("Enter your Zenodo API token: ")

BASE_URL = "https://zenodo.org/api"

def upload_mosei_to_zenodo(directory, token):
    """Upload all .csd files from directory to Zenodo"""

    print("="*70)
    print("UPLOADING TO ZENODO")
    print("="*70)

    # Create deposit
    print("\n[1/4] Creating deposit...")
    r = requests.post(f"{BASE_URL}/deposit/depositions",
                     params={"access_token": token}, json={})
    deposit = r.json()
    # Check for errors in the response before trying to access keys
    if r.status_code != 201: # 201 Created is the expected status for a new deposit
        print(f"API Error: {r.status_code} - {r.text}")
        # Attempt to get error details if available
        try:
            error_details = r.json()
            print(f"Error details: {error_details}")
        except json.JSONDecodeError:
            pass
        raise ValueError("Failed to create Zenodo deposit. Check your token and API call.")

    bucket = deposit["links"]["bucket"]
    deposit_id = deposit["id"]
    print(f"✓ Deposit ID: {deposit_id}")

    # Find and upload all .csd files
    print("\n[2/4] Uploading files...")
    csd_files = [f for f in os.listdir(directory) if f.endswith('.csd')]

    uploaded = []
    for csd_file in csd_files:
        filepath = os.path.join(directory, csd_file)
        size_gb = os.path.getsize(filepath) / (1024**3)
        print(f"  Uploading {csd_file} ({size_gb:.2f} GB)...", end=" ", flush=True)

        with open(filepath, "rb") as f:
            r = requests.put(f"{bucket}/{csd_file}",
                           data=f, params={"access_token": token})

        if r.status_code in [200, 201]:
            print("✓")
            uploaded.append(r.json())
        else:
            print(f"FAILED: {r.text}")
            raise ValueError(f"Failed to upload file {csd_file}: {r.text}")

    print(f"\n✓ Uploaded {len(uploaded)} files")

    # Add metadata
    print("\n[3/4] Adding metadata...")
    file_list = '\n'.join(f'- {f["key"]}' for f in uploaded)

    metadata = {
        "metadata": {
            "title": "MOSEI Processed Dataset - Multimodal Features",
            "upload_type": "dataset",
            "description": f"""Processed MOSEI dataset with multiple modalities.

**Files:**
{file_list}

**Usage:**
```python
from mmsdk import mmdatasdk
recipe = {{'Labels': 'DOWNLOAD_URL'}}
dataset = mmdatasdk.mmdataset(recipe)
```

Original: CMU-MOSEI (Zadeh et al., 2018)
SDK: CMU-MultimodalSDK
""",
            "creators": [{"name": "Your Name"}],
            "keywords": ["multimodal", "sentiment", "MOSEI", "audio", "visual", "text"],
            "access_right": "open",
            "license": "cc-by-4.0"
        }
    }

    r = requests.put(f"{BASE_URL}/deposit/depositions/{deposit_id}",
                    params={"access_token": token}, json=metadata)
    if r.status_code != 200: # 200 OK is expected for metadata update
        print(f"API Error: {r.status_code} - {r.text}")
        raise ValueError("Failed to add metadata to Zenodo deposit.")
    print("✓ Metadata added")

    # Publish
    print("\n[4/4] Publishing...")
    r = requests.post(f"{BASE_URL}/deposit/depositions/{deposit_id}/actions/publish",
                     params={"access_token": token})
    if r.status_code != 202: # 202 Accepted is expected for publish action
        print(f"API Error: {r.status_code} - {r.text}")
        raise ValueError("Failed to publish Zenodo deposit.")
    result = r.json()

    print("\n" + "="*70)
    print("✓✓✓ SUCCESS ✓✓✓")
    print("="*70)
    print(f"\nDOI: {result['doi']}")
    print(f"URL: {result['links']['record_html']}")

    print("\n📥 Download URLs:")
    print("-"*70)

# Run the upload
result = upload_mosei_to_zenodo("/content/filter_mosei", ZENODO_TOKEN)
````
